```python
n = int(input("cuantas materias tienes asignadas: "))
print(n)
lista_materias = []

for i in range(0,n):
    asignaturas = input("Nombre de asignaturas: ")
    lista_materias.append(asignaturas)
    
    for materia in lista_materias:
        # que nota tienes por asignatura, iteramos la lista de materias y eliminamos las materias pasadas.
        nota = float(input("Que nota tienes en "+ materia + "? "))
        if nota >= 60:
            lista_materias.remove(materia)
        
print("Debes retomar las siguientes materias: ", lista_materias)



    
```
