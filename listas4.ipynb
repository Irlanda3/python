```python
elementos = int(input("Teclea el numero de elementos a almacenar: "))

lista = []

# Estos numeros los voy almacenar en un list. Cuantas veces voy a iterar? Que numeros se almacenara?

for i in range(0, elementos):
    numeros = int(input())
    lista.append(numeros)
    lista.sort()
  
print(lista)
```
